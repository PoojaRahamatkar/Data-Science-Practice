```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
## Correlation with Heatmap
  # A Correlation heatmap uses colored cells, typically in a monochromatic scale, to show 2D correlation matrix(table) between two discrete dimensions or event types. 
  # It is very important in Feature Selection
```

```python

df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
# Joint Plot
  # A jointplot allows to study the relationship between 2 numeric variables.
  # The Central chart display their correlation.
  # It is usually a scatterplot, a haxbin, a 2D Histogram or a 2D Density Plot.
```

```python
## Univariate Analysis
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill', data=df, kind='reg')
```

Pair Plot

A "pairs plot" is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value. Like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

Dist Plot

Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False, bins=10)
```

Categorial Plots 

Seaborn also helps us in doing the analysis on Cateforial Data Points, In this section we will discuss about.
* Boxplot
*Violinplot
*countplot
*barplot

```python
# Count plot
sns.countplot('sex', data = df)
```

```python
# Count Plot
sns.countplot(y='sex', data=df)
```

```python
## Bar Plot
sns.barplot(x='total_bill', y='sex', data=df)
```

```python
## Bar plot
sns.barplot(x='sex', y='total_bill',data=df)
```

```python
df.head()
```

Box Plot

A box and whisker plot (Sometimes called a boxplot) is a graph that presents information form five-number summary.

```python
sns.boxplot('smoker', 'total_bill', data = df)
```

```python
sns.boxplot(x='day', y='total_bill', data = df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient ='v')
```

```python
# Categorize my data based on some other categories
sns.boxplot(x='total_bill',y='day', hue='smoker', data=df)
```

Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernal density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df, palette="rainbow")
```
