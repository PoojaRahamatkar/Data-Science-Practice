**Univariate, Biavariate and Multivariate Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

**Univariate Analysis**

Univariate analysis refers to the analysis of one variable. The prefix "Uni" means "one". 

There are three common ways to perform univariate analysis on one variable:

1. Summary Statistics :- Measures the center and spread of values.

2. Frequency table :- Describes how often different values occur.

3. Charts :- Used to visualize the distribution of values.

```python
df.loc[df['species']=='setosa']
```

```python
df_setosa = df.loc[df['species']=='setosa']
df_virginica = df.loc[df['species']=='virginica']
df_versicolor = df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']), 'o')
```

```python
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_virginica['sepal_length']), 'o')
plt.xlabel("Sepal Length")
plt.show()
```

```python
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']),  'o')
plt.xlabel("Sepal Length")
plt.show()
```

```python
plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']), 'o')
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_virginica['sepal_length']), 'o')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']), 'o')
plt.xlabel("Sepal Length")
plt.show()
```

**Bivariate Analysis**

Bivariate analysis is a kind of statistical analysis when two variables are observed against each other. One of the variables will be dependent and the other is independent. The variables are denoted by X and Y. The changes are analyzed between the two variables to understand to what extent the change has occurred.

```python
sns.FacetGrid(df, hue="species", size=5).map(plt.scatter, "sepal_length", "sepal_width").add_legend();
plt.show()
```

```python
sns.FacetGrid(df, hue="species", size=5).map(plt.scatter, "petal_length", 'petal_width').add_legend();
plt.show()
```

Note:- We have write down each and every observation when we plot a data or graph or observe the outliers of the data set for the future analysis.

**Multvariate Analysis**

Multivariate analysis is required when more than two variables have to be analyzed simultaneously. 

```python
sns.pairplot(df, hue="species", size=4)
```
